# 4. Đặt câu hỏi và trả lời

```python
#import các thứ viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#đọc file csv
df = pd.read_csv('data/mxmh_survey_results_clean.csv')
```

### Câu 1: Những người có mức độ bệnh tâm lý nặng thường có thói quen nghe nhạc như thế nào? Và việc này có giúp cải thiện tâm lý của họ không?

- Ở đây chúng ta xét mực độ nặng sẽ từ 6 - 10 

```python
df_anxiety = df[df['Anxiety'] > 5]
df_depression = df[df['Depression'] > 5]
df_insomnia = df[df['Insomnia'] > 5]
df_ocd = df[df['OCD'] > 5]

# vẽ biểu đồ cột số giờ nghe nhạc mỗi ngày của từng nhóm người trên
avg_hours_anxiety = df_anxiety['Hours per day'].mean()
avg_hours_depression = df_depression['Hours per day'].mean()
avg_hours_insomnia = df_insomnia['Hours per day'].mean()
avg_hours_ocd = df_ocd['Hours per day'].mean()

groups = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
avg_hours = [avg_hours_anxiety, avg_hours_depression, avg_hours_insomnia, avg_hours_ocd]

plt.barh(groups, avg_hours, color = 'green')
plt.title('Average hours of listening to music per day')
plt.xlabel('Hours')
plt.ylabel('Groups')
plt.show()
```

```python
print('Trung bình số giờ nghe nhạc của một người trong bài khảo sát là: ', df['Hours per day'].mean(), 'giờ')
```

- Có thể thấy trung bình số giờ nghe nhạc trong ngày của các nhóm người này cao hơn so với trung bình của tất cả mọi người

```python
#Vẽ biểu đồ frequency [gernes] của từng nhóm người trên
# plt, ax = plt.subplots(4, 1, figsize = (10, 10))
#Anxiety
df_anxiety_genres = df_anxiety.filter(like = 'Frequency')
avg_anxiety_genres = df_anxiety_genres.mean()
avg_anxiety_genres.plot(kind = 'barh', color = 'green')
plt.title('Average frequency of listening to music by genre of people with anxiety')
plt.xlabel('Frequency')
plt.ylabel('Genres')
plt.tight_layout()
plt.show()
#Depression
df_depression_genres = df_depression.filter(like = 'Frequency')
avg_depression_genres = df_depression_genres.mean()
avg_depression_genres.plot(kind = 'barh', color = 'green')
plt.title('Average frequency of listening to music by genre of people with depression')
plt.xlabel('Frequency')
plt.ylabel('Genres')
plt.tight_layout()
plt.show()
#Insomnia
df_insomnia_genres = df_insomnia.filter(like = 'Frequency')
avg_insomnia_genres = df_insomnia_genres.mean()
avg_insomnia_genres.plot(kind = 'barh', color = 'green')
plt.title('Average frequency of listening to music by genre of people with insomnia')
plt.xlabel('Frequency')
plt.ylabel('Genres')
plt.tight_layout()
plt.show()
#OCD
df_ocd_genres = df_ocd.filter(like = 'Frequency')
avg_ocd_genres = df_ocd_genres.mean()
avg_ocd_genres.plot(kind = 'barh', color = 'green')
plt.title('Average frequency of listening to music by genre of people with OCD')
plt.xlabel('Frequency')
plt.ylabel('Genres')
plt.show()
```

- Việc nghe nhạc như thế này có giúp cải thiện tinh thần của họ không?

```python
df_axiety_effect = df_anxiety['Music effects'].value_counts()
df_depression_effect = df_depression['Music effects'].value_counts()
df_insomnia_effect = df_insomnia['Music effects'].value_counts()
df_ocd_effect = df_ocd['Music effects'].value_counts()
#vẽ biểu đồ tròn tác động của nhạc đến tâm trạng của từng nhóm người trên
#Anxiety
df_axiety_effect.plot(kind = 'pie', autopct = '%.2f%%', figsize = (10, 10))
plt.title('Music effects on anxiety')
plt.show()
#Depression
df_depression_effect.plot(kind = 'pie', autopct = '%.2f%%', figsize = (10, 10))
plt.title('Music effects on depression')
plt.show()
#Insomnia
df_insomnia_effect.plot(kind = 'pie', autopct = '%.2f%%', figsize = (10, 10))
plt.title('Music effects on insomnia')
plt.show()
#OCD
df_ocd_effect.plot(kind = 'pie', autopct = '%.2f%%', figsize = (10, 10))
plt.title('Music effects on OCD')
plt.show()
```

Có thể nhận thấy phần lớn những người gặp các vấn đề về tinh thần ở mức độ nặng tham gia khảo sát đều cho rằng âm nhạc có thể giúp họ cải thiện tinh thần. Nhưng bên cạnh đó vẫn còn một số người cho rằng âm nhạc không có ảnh hưởng với họ và thậm chí còn làm cho tinh thần của họ trở nên tồi tệ đi

***Vậy nhịp tim có ảnh hưởng tới mức độ các vấn đề tâm lý của người tham gia khảo sát hay không? Liệu những người có nhịp tim cao (> 100) có nguy cơ mắc các vấn đề về tâm lý nặng hơn không?***

```python
df_high_bpm = df[df['BPM'] > 100]
df_low_bpm = df[df['BPM'] <= 100]
fig, ax = plt.subplots(2, 2, figsize = (20, 10))
ax[0, 0].barh(['High BPM', 'Low BPM'], [df_high_bpm['Anxiety'].mean(), df_low_bpm['Anxiety'].mean()])
ax[0, 0].set_title('Average anxiety of people with high and low BPM')
ax[0, 0].set_xlabel('Anxiety')
ax[0, 0].set_ylabel('BPM')
ax[0, 1].barh(['High BPM', 'Low BPM'], [df_high_bpm['Depression'].mean(), df_low_bpm['Depression'].mean()])
ax[0, 1].set_title('Average depression of people with high and low BPM')
ax[0, 1].set_xlabel('Depression')
ax[0, 1].set_ylabel('BPM')
ax[1, 0].barh(['High BPM', 'Low BPM'], [df_high_bpm['Insomnia'].mean(), df_low_bpm['Insomnia'].mean()])
ax[1, 0].set_title('Average insomnia of people with high and low BPM')
ax[1, 0].set_xlabel('Insomnia')
ax[1, 0].set_ylabel('BPM')
ax[1, 1].barh(['High BPM', 'Low BPM'], [df_high_bpm['OCD'].mean(), df_low_bpm['OCD'].mean()])
ax[1, 1].set_title('Average OCD of people with high and low BPM')
ax[1, 1].set_xlabel('OCD')
ax[1, 1].set_ylabel('BPM')
plt.show()
```

- Với những người có nhịp tim cao khi nghe dòng nhạc yêu thích thì mức độ về các vấn đề tâm lý của họ thường sẽ nặng hơn so với những người có nhịp tim thấp, nhưng chênh lệch là không quá lớn.

- Thử xem xét độ tương quan giữa nhịp tim và các giá trị này

Dòng nhạc nào phù hợp cho từng người bị mắc bệnh tâm lí cụ thể ?
- Lấy các khảo sát mà người thực hiện cảm thấy cải thiện tâm lí
- Voi cac muc do cua benh nho hon 8, xem xet voi gia tri trung binh

```python
df_improve=df[df['Music effects']=='Improve']
df_improve['mean_value'] = df_improve[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean(axis=1)
df_improve
```

```python
avg_anxiety=df_improve['Anxiety'].mean()
avg_depression=df_improve['Depression'].mean()
avg_insomnia=df_improve['Insomnia'].mean()
avg_OCD=df_improve['OCD'].mean()
df_anxiety = df_improve[(df_improve['Anxiety'] > avg_anxiety) | (df_improve['Anxiety'] > df_improve['mean_value'])]
df_depression = df_improve[(df_improve['Depression'] > avg_depression) | (df_improve['Depression'] > df_improve['mean_value'])]
df_insomnia = df_improve[(df_improve['Insomnia'] > avg_insomnia) | (df_improve['Insomnia'] > df_improve['mean_value'])]
df_OCD= df_improve[(df_improve['OCD'] > avg_OCD) | (df_improve['OCD'] > df_improve['mean_value'])]
df_anxiety
```

```python
df_anxiety_genres=df_anxiety.filter(like='Frequency').sum().to_frame().T
df_depression_genres=df_depression.filter(like='Frequency').sum().to_frame().T
df_insomnia_genres=df_insomnia.filter(like='Frequency').sum().to_frame().T
df_OCD_genres=df_OCD.filter(like='Frequency').sum().to_frame().T
def bar_plot(df, title):
    fig, ax = plt.subplots(figsize=(20,5))
    df = df.sum()
    df = df.sort_values(ascending=True)
    colors = plt.cm.viridis(np.linspace(0, 1, len(df)))

    ax = df.plot(kind='barh', color=colors)
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
    ax.set_title(title)
    for i, v in enumerate(df):
        ax.text(v, i, str(v),  va='center')

    plt.show()

bar_plot(df_anxiety_genres, 'Anxiety')
bar_plot(df_depression_genres, 'Depression')
bar_plot(df_insomnia_genres, 'Insomnia')
bar_plot(df_OCD_genres, 'OCD')
```
